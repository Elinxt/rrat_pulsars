```python
def prec(a,d,t0):
    rg=1.7453292e-2
    rm=2.90888e-4
    rs=4.8481368e-6
    am=46.1            
    an=20.0
    am=am*rs
    an=an*rs
    alfa=(a[0]*rg+a[1]*rm+a[2]*rs)*15 
    delta=(d[0]*rg+d[1]*rm+d[2]*rs)
    t=((t0)-2000)
    alfa1=int(alfa)
    delta1=int(delta)
    for i in range (2):
            alf = (alfa + (am + an * math.sin(alfa1) * math.tan(delta1)) * t)
            dell = (delta + an * math.cos(alfa1) * t)
            alfa1 = ((alf + alfa) / 2)
            delta1 = ((dell + delta) / 2)
    alf=(alf/(15*rs)) 
    dell=(dell/rs) #del в фортр
    n = []
    m = []
    for i in range (3):
            n.append(i)
            m.append(i)
    m[0]=int(alf/3600) #допустим int
    m[1]=int((alf/3600-m[0])*60)
    m[2]=int((((alf/3600-m[0])*60)-m[1])*60 + 0.5) #+0.5
    n[0]=int(dell/3600)
    n[1]=int((dell/3600-n[0])*60)
    n[2]=int((((dell/3600-n[0])*60)-n[1])*60 + 0.5)  #+0.5
    if m[0] >= 24:
        m[0] = m[0]-24
        year = t0     #что это?
    a[0] = m[0]
    a[1] = m[1]
    a[2] = m[2]
    d[0] = n[0]
    d[1] = n[1]
    d[2] = n[2]
    return a, d, t0
```
